```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data=np.array(["Male","Female","Male","Male","Female"])
```

```python
plt.Figure(figsize=(3,2))
sns.countplot(x=data)
plt.yticks([1,2,3,4])
plt.title("Count of gender")
plt.show()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
plt.Figure(figsize=(6,4))
sns.countplot(tips,x="day",hue="smoker",palette="magma")
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.barplot(tips,x="day",y="tip",estimator="sum",hue="sex",palette="rocket")
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.histplot(tips,x="total_bill",hue="sex",palette=["r","b"],multiple="stack",kde=True,log_scale=True)
plt.show()
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
plt.Figure(figsize=(6,4))
sns.histplot(iris,x="petal_length",hue="species",palette="magma",multiple="stack",kde=True,bins=20)
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.displot(data=iris,x="petal_width",hue="species",kind="kde",multiple="stack",palette="magma")
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.boxplot(data=tips,x="day",y="total_bill",hue="smoker",palette=["m","g"],saturation=0.99)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")


plt.Figure(figsize=(6,4))
sns.scatterplot(data=iris,x="petal_length",y="petal_width",hue="species",size="sepal_length",marker="*",palette="magma")
plt.show()
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip",kind="hex",height=5)
plt.show()
```

```python
sns.jointplot(data=iris,x="petal_width",y="petal_length",kind="kde",height=5,hue="species")
plt.show()
```

```python
sns.lineplot(data=tips,x="total_bill",y="tip",hue="smoker",dashes=False)
plt.show()
```

```python
sns.pairplot(data=iris,hue="species",palette="rocket",kind="kde")
plt.show()
```

```python
sns.violinplot(data=iris,x="sepal_length",hue="species")
plt.show()
```
