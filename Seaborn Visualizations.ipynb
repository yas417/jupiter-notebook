```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = np.array(["Male", "Female", "Male", "Male", "Female", "Male", "Female"])
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x=data)
plt.yticks([1, 2, 3, 4])
plt.title("Count of Gender")
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(tips, x="day", hue="smoker", palette="magma")
plt.show()
```

```python
tip_day = tips.groupby("day")["tip"].sum()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(tips, x="day", y="tip", estimator="sum", hue="sex",
           palette="rocket")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(tips, x="day", y="tip", estimator="mean", hue="sex",
           palette="rocket")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(tips, x="total_bill", hue="sex", palette=["r", "b"],
            multiple="stack", kde=True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(tips, x="total_bill", hue="day", palette="light:m_r",
            multiple="stack", kde=True)
plt.show()
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(iris, x="petal_length", hue="species", palette="magma",
            multiple="stack", kde=True, bins=20)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.displot(data=iris, x="petal_width", hue="species", kind="kde", multiple="stack",
           palette="dark:salmon_r")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker", palette=["m", "g"],
           saturation=0.99)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(tips, x="total_bill", y="tip", hue="sex", size="size",marker="*")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(iris, x="petal_length", y="petal_width", hue="species", 
                size="sepal_length",marker="*", palette="magma")
plt.show()
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", height=5, kind="scatter",hue="smoker")
plt.show()
```

```python
sns.jointplot(data=iris, x="petal_width", y="petal_length",height=5, kind="kde",
              hue="species", palette="rocket")
plt.show()
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip", hue="smoker", dashes=False)
```

```python
sns.pairplot(data=iris, hue="species", palette="rocket", kind="kde")
```

```python
sns.violinplot(data=tips, x="total_bill", y="day")
```

```python
sns.violinplot(data=iris, x="sepal_length", hue="species")
```
